```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
url='https://www.sportskeeda.com/go/pro-kabaddi'
```

```python
driver=webdriver.Chrome(r'C:\Users\Administrator\Downloads\chromedriver_win32\chromedriver.exe')
driver.get(url)
driver.maximize_window()
element=driver.find_element_by_link_text("Results")
element.click()
```

```python
soup1=BeautifulSoup(driver.page_source)
```

```python
previous_results=soup1.find_all('a',{'class':'one-table-row'})
```

```python
len(previous_results)
```

```python
dis={}
date=[]
team=[]
score1=[]
opponent=[]
score2=[]
```

```python
for i in range(len(previous_results)):
    date.append(previous_results[i].find_all('div',{'class':'match-date'})[0].text)
    team.append(previous_results[i].find_all('div',{'class':'team-player-name'})[0].text)
    score1.append(previous_results[i].find_all('div',{'class':'team-score'})[0].text)
    opponent.append(previous_results[i].find_all('div',{'class':'team-player-name'})[1].text)
    score2.append(previous_results[i].find_all('div',{'class':'team-score'})[1].text)
dis['date']=date
dis['team']=team
dis['opponent']=opponent
dis['score1']=score1
dis['score2']=score2
```

```python
previous_result_df=pd.DataFrame(dis,index=range(len(previous_results)))
```

```python
def winner(row):
    date,t,o,score_team,score_opponent=row
    if int(score_team)>int(score_opponent):
        return t
    elif int(score_team)==int(score_opponent):
        return "Tie"
    else:
        return o
```

```python
previous_result_df['Winner']=previous_result_df.apply(winner,axis=1)
```
